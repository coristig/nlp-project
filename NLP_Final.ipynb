```python
from urllib import request
import time
import pandas as pd
import nltk
```

```python
df = pd.read_csv('authors.csv')
```

```python
def get_text(x):
    response = request.urlopen(x)
    raw = response.read().decode('utf8')
    time.sleep(2.5)
    return raw
```

```python
df['text'] = df['URL'].map(get_text)
```

```python
df
```

```python
def cleaned(item):
    if item.find("CONTENTS")!=-1:
        start = item.find("CONTENTS")
    else: start=0
    if item.find("End of the Project Gutenberg")!=-1:
        end = item.find("End of the Project Gutenberg")
    else: end = 0
    return item[start:end] 
```

```python
df['text_clean'] = df['text'].map(cleaned)
```

```python
tree_test = df['text_clean'][3]
tree_test_sent = nltk.sent_tokenize(tree_test)
tree_test_w = [nltk.word_tokenize(sentence) for sentence in tree_test_sent]

tree_pos = [nltk.pos_tag(sentence) for sentence in tree_test_w]
```

```python
def calculate_pos_probability(text):
    sent_tree = nltk.sent_tokenize(text)
    word_tree = [nltk.word_tokenize(sentence) for sentence in sent_tree]

    pos_tree = [nltk.pos_tag(word) for word in word_tree]
    freqs = {}
    for sent in pos_tree:
        last_pos = None
        for word, pos in sent:
            if last_pos is None:
                last_pos = pos
                continue
            else:
                if last_pos not in freqs:
                    freqs[last_pos] = {}
                freqs[last_pos][pos] = freqs[last_pos].get(pos, 0) + 1
                last_pos = pos
    return freqs
pos_probs = calculate_pos_probability("Greg walked the dog. Then Greg ate the dog.")
pos_probs
```

```python
pos_probs = calculate_pos_probability(df.text_clean[0])
```

```python
pos_probs
```

```python
for pos in pos_probs:
    pos_keys = sorted(pos_probs[pos], key=lambda x: pos_probs[pos][x], reverse=True)
    for next_pos in pos_keys[:5]:
        freq = pos_probs[pos][next_pos] / sum(pos_probs[pos].values())
        print(pos, next_pos, pos_probs[pos][next_pos], freq)
    print("*"*25)
```
